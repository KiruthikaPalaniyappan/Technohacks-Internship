```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/content/Iris.csv', header=None)
iris.columns = ['id', 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
plt.hist(iris['sepal_length'], bins=34)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(iris['sepal_width'], bins=34)
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(iris['petal_length'], bins=20)
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(iris['petal_width'], bins=18)
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```
